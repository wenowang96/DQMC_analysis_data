```python
%reload_ext autoreload
%autoreload 2

from packages.correlations import *
from packages.modelfunction import *
from packages.correlation_length import *
```

```python
def save_correlationlength(U,sitex,sitey,tp,betas,END=None,START=2):
    correlationlengths =  np.zeros((len(betas)))
    zz = np.load(f'zzreal_{sitex:g}*{sitey:g}U{U:g}_tp{tp:g}.npy')
    for j, b in enumerate(betas):
        print(b)
        correlationlengths[j] = correlation_length_alongx(zz[j],PLOT=True,start=START,end=END)
        np.save(f'correlationlength_{sitex:g}*{sitey:g}U{U:g}_tp{tp:g}.npy',correlationlengths)
```

```python
U=10
sitex = 6
sitey = 6
tp=0

betas = np.array([2,2.5,3,3.5,4,4.5,5,6,7,8])

save_correlationlength(U,sitex,sitey,tp,betas)
```
