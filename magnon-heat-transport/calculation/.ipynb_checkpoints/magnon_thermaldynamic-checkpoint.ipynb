```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def omega(J,kx,ky):
    z = 2*2
    s=1/2
    gamma = 2/z*(np.cos((kx+ky)/2)+np.cos((kx-ky)/2))   
    return J*s*z*np.sqrt(1-gamma**2)
```

```python
def records(J,betas):
    cs = np.zeros(len(betas))
    for j,beta in enumerate(betas[:]): 
        def cf(kx,ky):
            Omega = omega(J,kx,ky)
            f = Omega**2*beta**2/(np.exp(Omega*beta)-1)**2 * np.exp(Omega*beta) *2 # this is for the two exitation

            return f

        c =integrate.nquad(cf, [[0, np.pi],[0, np.pi]])[0]*4 / (4*np.pi**2)  

        cs[j]=c
        print(beta,1/beta,c)


        np.save(f"magnon_cs_U{U:g}_continue.npy", cs)
```

```python
betas = 1/np.linspace(0.05,0.25,100)
U=8
t=1

J=4*t**2/U
records(J,betas)
```

```python
def records_cvsquare(J,betas):
    cs = np.zeros(len(betas))
    for j,beta in enumerate(betas[:]): 
        z = 2*2
        s= 1/2
        def cf(kx,ky):
            Omega = omega(J,kx,ky)
            gamma = 2/z*(np.cos((kx+ky)/2)+np.cos((kx-ky)/2))   
            f = Omega**2*beta**2/(np.exp(Omega*beta)-1)**2 * np.exp(Omega*beta) *2 # this is for the two exitation
            partialgammak = 2/z*(np.sin((kx+ky)/2)+np.sin((kx-ky)/2))/np.sqrt(2)
            f *= ((J*z*s)**2/Omega*gamma*(partialgammak))**2     #v_x(k)^2
            return f

        c =integrate.nquad(cf, [[0, np.pi],[0, np.pi]])[0]*4 / (4*np.pi**2)

        cs[j]=c
        print(beta,c)


        np.save(f"magnon_cvksquares_U{U:g}_continue.npy", cs)
```

```python
betas = 1/np.linspace(0.05,0.25,100)
U=10
t=1

J=4*t**2/U
records_cvsquare(J,betas)
```

```python
def records_cvv(J,betas):
    cs = np.zeros(len(betas))
    for j,beta in enumerate(betas[:]): 
        z = 2*2
        s= 1/2
        def cfv(kx,ky):
            Omega = omega(J,kx,ky)
            gamma = 2/z*(np.cos((kx+ky)/2)+np.cos((kx-ky)/2))   
            f = Omega**2*beta**2/(np.exp(Omega*beta)-1)**2 * np.exp(Omega*beta) *2 # this is for the two exitation
            partialgammak = 2/z*(np.sin((kx+ky)/2)+np.sin((kx-ky)/2))/np.sqrt(2)
            f *= (J*z*s)**2/Omega*gamma*(partialgammak)  #v_x(k)^2
            return f
        def cf(kx,ky):
            Omega = omega(J,kx,ky)
            gamma = 2/z*(np.cos((kx+ky)/2)+np.cos((kx-ky)/2))   
            f = Omega**2*beta**2/(np.exp(Omega*beta)-1)**2 * np.exp(Omega*beta) *2 # this is for the two exitation
            #partialgammak = 2/z*(np.sin((kx+ky)/2)+np.sin((kx-ky)/2))/np.sqrt(2)
            #f *= (J*z*s)**2/Omega*gamma*(partialgammak)    #v_x(k)^2
            return f

        cvv =integrate.nquad(cfv, [[0, np.pi],[0, np.pi]])[0]*4 / (4*np.pi**2)
        cv = integrate.nquad(cf, [[0, np.pi],[0, np.pi]])[0]*4 / (4*np.pi**2)


        
        cs[j]=cvv/cv
        print(beta,cs[j])


        np.save(f"magnon_cvv_U{U:g}_continue.npy", cs)
```

```python
betas0 = np.array([0.125,0.25,0.3,0.4,0.5,0.6,0.7,1,1.5,2,2.5,3,3.5,4,4.5,5,6,7,8,10,12,20])
betas = 1/((1/betas0[:-1]+1/betas0[1:])/2)
U=10
t=1

J=4*t**2/U
records_cvv(J,betas)
```
