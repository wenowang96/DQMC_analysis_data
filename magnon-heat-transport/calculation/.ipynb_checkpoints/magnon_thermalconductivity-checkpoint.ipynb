```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx1 = (np.abs(array - value)).argmin()
    idx = 0
    while array[idx+1]>=array[idx] and (idx+1)<len(array)-1:
        idx +=1;
    #go the the first peak
    
    while idx<(idx1+1) and array[idx+1]>value and array[idx+1]<=array[idx]:
        idx+=1
        
    return idx
```

```python
def loading(U,tp,sitex,sitey,prefix):
    jeje = np.load("../calculation/" + prefix+f"jejeU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    jkjk = np.load("../calculation/" +prefix+f"jkjkU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    jpjp = np.load("../calculation/" +prefix+f"jpjpU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    j0j0 = np.load("../calculation/" +prefix+f"j0j0U{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    
    C = np.load("../calculation/" +prefix+f"CU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    chi = np.load("../calculation/" +prefix+f"chiU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    
    jejes = np.load("../calculation/" +prefix+f"jejesU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    jkjks = np.load("../calculation/" +prefix+f"jkjksU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    jpjps = np.load("../calculation/" +prefix+f"jpjpsU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    j0j0s = np.load("../calculation/" +prefix+f"j0j0sU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    
    Cs = np.load("../calculation/" +prefix+f"CsU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    chis = np.load("../calculation/" +prefix+f"chisU{U:g}tp{tp:g}_{sitex:g}*{sitey:g}.npy")
    
    return jeje ,jkjk,jpjp,j0j0,C,chi,jejes,jkjks,jpjps,j0j0s,Cs,chis

def fitfun_drude(w,A,gamma):
    return A/(gamma**2+w**2)

```

```python
def calc(betas,U,tp,sitex,sitey,prefix = "transport_"):
    Ajeje,Ajkjk,Ajpjp,Aj0j0,C,chi,jejes,jkjks,jpjps,j0j0s,Cs,chis = \
    loading(U,tp,sitex,sitey,prefix)
    w = np.load(f"wU={U:g}.npy")
    dw = np.load(f"dwU={U:g}.npy")
    Ds = np.zeros((len(betas),2))
    taus = np.zeros((len(betas),2))
    for j,beta in enumerate(betas[:]):
        k=0
        jeje,jkjk,jpjp,j0j0 = Ajeje[k,j],Ajkjk[k,j],Ajpjp[k,j],Aj0j0[k,j]
        ks =(jeje[:])*beta
        kappa1part1 = ( jkjk[:])*beta
        kappa2part1  = ( jpjp[:])*beta
        part2 = (ks-kappa1part1-kappa2part1)/2
        
        tau = 1/w[find_nearest((kappa1part1+part2), (kappa1part1+part2)[0]/2)]
        taus[j,0]=tau
        Ds[j,0]=np.sum((kappa1part1+part2)[:200]*dw[:200])
        
        plt.plot(w[:200],(kappa1part1+part2)[:200],color='b')
        plt.plot([tau**(-1),tau**(-1)],[0,(kappa1part1+part2)[0]*1.1],color='r')
        plt.ylim(-0.5*(kappa1part1+part2)[0],(kappa1part1+part2)[0]*1.1)
        plt.axhline(y=0)
        plt.show()
        
        jejesamples,jkjksamples,jpjpsamples,j0j0samples = \
        jejes[k,j],jkjks[k,j],jpjps[k,j],j0j0s[k,j]
        
        Dsamples = np.zeros(len(jejesamples[:,0]))
        ks =(jejesamples)*beta
        kappa1part1 = ( jkjksamples)*beta
        kappa2part1  = ( jpjpsamples)*beta
        part2 = (ks-kappa1part1-kappa2part1)/2
        for i in range(len(jejesamples[:,0])):
            Dsamples[i] = np.sum((kappa1part1+part2)[i,:200]*dw[:200])
        print(w[200])
        Ds[j,1]= np.std(Dsamples,ddof = 1)
        tausamples = np.zeros(len(jejesamples[:,0]))
        for i in range(len(jejesamples[:,0])):
            tausamples[i] = 1/w[find_nearest((kappa1part1+part2)[i], (kappa1part1+part2)[i,0]/2)]
        taus[j,1]= np.std(tausamples,ddof = 1)
        
        print(beta)
        
    np.save(f"Drudeweight_U{U:g}.npy", Ds)
    np.save(f"relaxationtime_U{U:g}.npy", taus)
```

```python
U=10
tp = 0
betasU10n1 = np.array([0.125,0.25,0.3,0.4,0.5,0.6,0.7,1,1.5,2,2.5,3,3.5,4,4.5,5,6,7,8,10,12,20])
prefixn1 = "n1_annealing_single_alwaysBT"
sitex = 8
sitey = 8

calc(betasU10n1,U,tp,sitex,sitey,prefix = prefixn1)
```
